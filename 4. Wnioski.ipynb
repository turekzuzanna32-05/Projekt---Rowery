W tym rozdziale podsumowujemy najważniejsze wyniki przeprowadzonej analizy, odnosimy się do postawionych pytań badawczych, wskazujemy ograniczenia badania oraz proponujemy kierunki dalszych analiz.

---

## 4.1 Podsumowanie wyników

Na podstawie przeprowadzonej analizy danych klientów sklepu rowerowego, odpowiadamy na postawione we wstępie pytania badawcze:

### 1. Jaki jest profil demograficzny typowego nabywcy roweru?

Typowy nabywca roweru to:

- Wiek: 30-50 lat (korelacja -0.11 - bardzo słaby wpływ)
- Dochód: brak wyraźnego związku (korelacja 0.05)
- Płeć: częściej mężczyźni (z wykresu słupkowego)
- Wykształcenie: wyższe (Graduate Degree lub Bachelors)
- Region: najczęściej Pacific
- Najważniejszy czynnik: liczba posiadanych samochodów (korelacja -0.20) - im mniej samochodów, tym większa szansa na zakup roweru!

### 2. Które czynniki najsilniej wpływają na decyzję o zakupie?

Największe znaczenie ma liczba posiadanych samochodów – im mniej samochodów, tym większa skłonność do zakupu roweru. Istotnym czynnikiem jest również wiek – najchętniej rowery kupują osoby w przedziale 30–50 lat. Wykształcenie wyższe oraz zamieszkanie w regionie Pacific również zwiększają prawdopodobieństwo zakupu. Płeć ma znaczenie drugorzędne, a dochód nie wykazuje wyraźnego związku z decyzją zakupową.

Powyższa skala opiera się na analizie porównawczej grup. Niskie współczynniki korelacji liniowej, widoczne na heatmapie, wynikają z nieliniowego charakteru danych demograficznych – np. rowery są najczęściej kupowane przez osoby w konkretnym przedziale wiekowym (30–50 lat), a nie w sposób liniowy wraz ze wzrostem wieku.

### 3. Czy istnieją różnice w preferencjach zakupowych między grupami?
Tak, zaobserwowano istotne różnice:

**Wiek:** największy odsetek kupujących występuje w grupie 30–50 lat, podczas gdy seniorzy (65+) rzadziej decydują się na zakup.

**Wykształcenie:** osoby z wyższym wykształceniem (Graduate Degree, Bachelors) kupują rowery znacznie częściej niż osoby z wykształceniem podstawowym lub średnim.

**Region:** mieszkańcy Pacific wykazują największe zainteresowanie rowerami, wyprzedzając Europę i Amerykę Północną.

**Stan cywilny:** osoby samotne częściej kupują rowery niż osoby w związkach małżeńskich.

**Płeć:** mężczyźni nieco częściej dokonują zakupu niż kobiety.

## 4.2 Ograniczenia badania

**Źródło danych** - Dane pochodzą z fikcyjnego sklepu rowerowego, co oznacza, że nie można ich uogólniać na cały rynek ani branżę rowerową. Wyniki odzwierciedlają specyfikę jednego, hipotetycznego biznesu.

**Brak informacji o cenach i promocjach** - W zbiorze danych zabrakło zmiennych dotyczących cen rowerów oraz dostępnych promocji. Są to czynniki, które mogą mieć istotny wpływ na decyzję zakupową, a ich pominięcie ogranicza pełne zrozumienie zachowań klientów.

**Brak danych o sezonowości** - Nie wiadomo, w której porze roku dokonano zakupu. Możliwe, że rowery są kupowane częściej wiosną lub latem, co mogłoby wpłynąć na interpretację wyników.

**Brak historii zakupów** - Zbiór danych nie zawiera informacji o tym, czy klient kupił rower po raz pierwszy, czy wymienia stary model. Rozróżnienie tych grup mogłoby ujawnić inne wzorce decyzyjne.

**Obecność wartości odstających** - W zbiorze występują wartości odstające, takie jak wiek 89 lat czy bardzo wysokie dochody. Choć nie zostały usunięte, mogą one zaburzać wyniki analiz statystycznych.

**Brak danych jakościowych** - Nie znamy powodów, dla których klienci zdecydowali się na zakup roweru – czy była to rekreacja, dojazdy do pracy, sport, czy chęć zmiany stylu życia. Tego typu informacje mogłyby wzbogacić analizę i pomóc w lepszym zrozumieniu motywacji klientów.

## 4.3 Rekomendacje dla sklepu

### A) Targetowanie klientów
1. **Grupa docelowa:** osoby w wieku 30-50 lat, z dochodami powyżej 70 000 USD
2. **Edukacja:** kampanie kierować do osób z wyższym wykształceniem
3. **Region:** zwiększyć nakłady marketingowe w regionie Pacific

Musimy jednak pamiętać, że wyniki badania wskazują na wyraźny profil demograficzny nabywcy (wiek 30-50 lat, wyższe dochody), jednakże brak silnych korelacji liniowych sugeruje, że sam dochód czy wiek nie są jedynymi 'wyzwalaczami' zakupu. Decyzja o nabyciu roweru jest wypadkową wielu czynników, co oznacza, że w strategii marketingowej nie powinniśmy polegać tylko na jednym parametrze, a raczej na segmentacji uwzględniającej wielowymiarowy profil klienta.

### B) Strategie promocyjne
1. **Program lojalnościowy** - dla klientów z wyższym wykształceniem
2. **Promocje sezonowe** - skierowane do młodszych klientów (30-40 lat)
3. **Oferty rodzinne** - zauważono, że posiadanie dzieci nie wpływa negatywnie na zakup

### C) Dalsze działania
1. Zebrać dane o cenach rowerów i preferencjach klientów
2. Przeprowadzić badanie satysfakcji klientów
3. Rozważyć modele predykcyjne do prognozowania zakupów

## 4.4 Zakończenie

Przeprowadzona eksploracyjna analiza danych pozwoliła na identyfikację kluczowych czynników wpływających na decyzję o zakupie roweru. Najważniejszymi z nich okazały się: **dochód, wiek, wykształcenie oraz region zamieszkania**.

Uzyskane wyniki mogą stanowić podstawę do:
- **Planowania kampanii marketingowych** skierowanych do odpowiednich grup klientów
- **Optymalizacji oferty** produktowej i promocyjnej
- **Budowy modeli predykcyjnych** wspierających proces sprzedaży
- **Weryfikacji polityki cenowej** wspierających proces sprzedaży
