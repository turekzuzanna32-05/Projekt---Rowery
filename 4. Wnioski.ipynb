<h2 id="wnioski">4. Wnioski</h2>

W tym rozdziale podsumowujemy najważniejsze wyniki przeprowadzonej analizy, odnosimy się do postawionych pytań badawczych, wskazujemy ograniczenia badania oraz proponujemy kierunki dalszych analiz.

## 4.1 Podsumowanie wyników

Na podstawie przeprowadzonej analizy danych klientów sklepu rowerowego, odpowiadamy na postawione we wstępie pytania badawcze:

### 1. Jaki jest profil demograficzny typowego nabywcy roweru?

Typowy nabywca roweru to:
- **Wiek:** 30-50 lat (średnia ok. 44 lata)
- **Dochód:** powyżej 70 000 USD (średnia ok. 74 000 USD)
- **Płeć:** częściej mężczyźni
- **Wykształcenie:** wyższe (Graduate Degree lub Bachelors)
- **Region:** najczęściej Pacific

### 2. Które czynniki najsilniej wpływają na decyzję o zakupie?

| Czynnik | Wpływ na zakup | Względne znaczenie czynnika |
|---|---|---|
| **Dochód** | Im wyższy, tym większa szansa na zakup | ★★★★★ |
| **Wiek** | Osoby w średnim wieku (30-50 lat) wykazują najwyższą skłonność do zakupu | ★★★★☆ |
| **Wykształcenie** | Wykształcenie wyższe dominuje w grupie nabywców | ★★★★☆ |
| **Region** | Pacific > Europa > Ameryka Płn. | ★★★☆☆ |
| **Płeć** | Mężczyźni częściej kupują | ★★☆☆☆ |

Zastosowana skala, czyli gwiazdki opiera się na analizie porównawczej grup (różnice w średnich i rozkładach częstotliwości). Niskie współczynniki korelacji liniowej widoczne na heatmapie w rozdziale 2.Wizualizacja wynikają ze specyfiki danych demograficznych. Są one często nieliniowe (np. rower jest kupowany najczęściej przez osoby w konkretnym "oknie" wiekowym 30-50 lat, a nie w sposób liniowy wraz ze wzrostem wieku, osoba w wieku lat 80 czy 20 nie kupi tak często roweru jak w średnim wieku). 

Dlatego do oceny siły wpływu przyjęłyśmy segmentację grupową, która w tym przypadku lepiej oddaje realne wzorce zakupowe niż proste korelacje, którą policzyłyśmy wcześniej. 


### 3. Czy istnieją różnice w preferencjach zakupowych między grupami?

Tak, zaobserwowano istotne różnice:
- **Dochodowe:** osoby zamożniejsze (powyżej 70 000 USD) stanowią główną grupę nabywców
- **Wiekowe:** osoby młodsze (30-50 lat) częściej kupują niż seniorzy
- **Edukacyjne:** wykształcenie wyższe silnie koreluje z zakupem
- **Regionalne:** mieszkańcy regionu Pacific wykazują największe zainteresowanie rowerami

## 4.2 Ograniczenia badania

Przedstawiona analiza ma pewne ograniczenia, które należy wziąć pod uwagę przy interpretacji wyników:

| Ograniczenie | Opis |
|---|---|
| **Jedno źródło danych** | Dane pochodzą z fikcyjnego sklepu rowerowego, nie można ich uogólniać na cały rynek |
| **Brak cen rowerów** | Brak informacji o cenach rowerów i promocjach, co mogłoby wpłynąć na decyzję zakupową |
| **Brak sezonowości** | Dane nie zawierają informacji o porze roku, w której dokonano zakupu |
| **Brak historii zakupów** | Nie wiemy, czy klient kupił rower po raz pierwszy, czy wymienia stary |
| **Wartości odstające** | Wiek 89 lat oraz bardzo wysokie dochody mogą zaburzać wyniki |
| **Brak danych jakościowych** | Nie znamy powodów zakupu (np. rekreacja, dojazdy do pracy, sport) |

## 4.3 Rekomendacje dla sklepu

Na podstawie wyników analizy, proponujemy następujące działania marketingowe:

### Targetowanie klientów
1. **Grupa docelowa:** osoby w wieku 30-50 lat, z dochodami powyżej 70 000 USD
2. **Edukacja:** kampanie kierować do osób z wyższym wykształceniem
3. **Region:** zwiększyć nakłady marketingowe w regionie Pacific

Musimy jednak pamiętać, że wyniki badania wskazują na wyraźny profil demograficzny nabywcy (wiek 30-50 lat, wyższe dochody), jednakże brak silnych korelacji liniowych sugeruje, że sam dochód czy wiek nie są jedynymi 'wyzwalaczami' zakupu. Decyzja o nabyciu roweru jest wypadkową wielu czynników, co oznacza, że w strategii marketingowej nie powinniśmy polegać tylko na jednym parametrze, a raczej na segmentacji uwzględniającej wielowymiarowy profil klienta.

Ten profil, co prezentujemy zdarza się po prostu najczęściej.

### Strategie promocyjne
1. **Program lojalnościowy** dla klientów z wyższym wykształceniem
2. **Promocje sezonowe** skierowane do młodszych klientów (30-40 lat)
3. **Oferty rodzinne** - zauważono, że posiadanie dzieci nie wpływa negatywnie na zakup

### Dalsze działania
1. Zebrać dane o cenach rowerów i preferencjach klientów
2. Przeprowadzić badanie satysfakcji klientów
3. Rozważyć modele predykcyjne do prognozowania zakupów

## 4.4 Zakończenie

Przeprowadzona eksploracyjna analiza danych pozwoliła na identyfikację kluczowych czynników wpływających na decyzję o zakupie roweru. Najważniejszymi z nich okazały się: **dochód, wiek, wykształcenie oraz region zamieszkania**.

Uzyskane wyniki mogą stanowić podstawę do:
- **Planowania kampanii marketingowych** skierowanych do odpowiednich grup klientów
