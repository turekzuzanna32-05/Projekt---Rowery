## 3. Wizualizacja

W tej sekcji przedstawiamy 5 wykresów ilustrujących zależność między cechami klientów a decyzją o zakupie roweru. Każdy wykres został opatrzony krótkim opisem i interpretacją. Zacznijmy od instalacji bibliotek i wczytania danych.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('data/sklep_rowerowy_oczyszczone.csv')
df.head()
```

**Wykres 1. Dochód a zakup roweru**

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x='Purchased Bike', y='Income', data=df, hue='Purchased Bike', palette='Set2', legend=False)
plt.title('Rozkład dochodów a zakup roweru', fontsize=14, fontweight='bold')
plt.xlabel('Zakup roweru', fontsize=12)
plt.ylabel('Dochód [USD]', fontsize=12)
plt.xticks([0, 1], ['Nie kupił', 'Kupił'])
plt.grid(axis='y', alpha=0.3)
plt.show()
```

Osoby, które kupiły rower, mają wyraźnie wyższe dochody (mediana ok. 80 000 USD) niż osoby, które nie dokonały zakupu (mediana ok. 40 000 USD). Widać również, że w grupie kupujących występuje większe zróżnicowanie dochodów.

**Interpretacja:** Dochód jest istotnym czynnikiem wpływającym na decyzję o zakupie roweru. Osoby zamożniejsze częściej decydują się na ten wydatek, co sugeruje, że rower jest postrzegany jako dobro luksusowe lub inwestycja w zdrowie.

**Wykres 2. Wiek a zakup roweru**

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x='Purchased Bike', y='Age', data=df, hue='Purchased Bike', palette='Set2', legend=False)
plt.title('Rozkład wieku a zakup roweru', fontsize=14, fontweight='bold')
plt.xlabel('Zakup roweru', fontsize=12)
plt.ylabel('Wiek [lata]', fontsize=12)
plt.xticks([0, 1], ['Nie kupił', 'Kupił'])
plt.grid(axis='y', alpha=0.3)
plt.show()
```

Kupujący rowery są przeciętnie młodsi (mediana ok. 40 lat) niż osoby niekupujące (mediana ok. 48 lat). W grupie kupujących wiek jest bardziej skupiony wokół mediany, podczas gdy w grupie niekupujących występuje szersze rozproszenie.

**Interpretacja:** Młodsze osoby (30-50 lat) częściej decydują się na zakup roweru. Może to wynikać z większej świadomości zdrowotnej, aktywnego trybu życia lub mniejszej liczby zobowiązań finansowych w porównaniu do starszych osób.

**Wykres 3. Płeć a zakup roweru**

```python
plt.figure(figsize=(8, 6))
sns.countplot(x='Gender', hue='Purchased Bike', data=df, palette='Set2')
plt.title('Płeć a zakup roweru', fontsize=14, fontweight='bold')
plt.xlabel('Płeć', fontsize=12)
plt.ylabel('Liczba osób', fontsize=12)
plt.legend(title='Zakup roweru', labels=['Nie kupił', 'Kupił'])
plt.grid(axis='y', alpha=0.3)
plt.show()
```

Mężczyźni częściej kupują rowery niż kobiety. Wśród mężczyzn liczba kupujących (ok. 280) jest zbliżona do liczby niekupujących (ok. 260), podczas gdy wśród kobiet dominują osoby, które nie dokonały zakupu (ok. 290) nad kupującymi (ok. 170).

**Interpretacja:** Istnieje różnica w preferencjach zakupowych między płciami. Mężczyźni są bardziej skłonni do zakupu roweru, co może wynikać z większego zainteresowania sportem lub różnic w stylu życia.

**Wykres 4. Region a zakup roweru**

```python
plt.figure(figsize=(10, 6))
sns.countplot(x='Region', hue='Purchased Bike', data=df, palette='Set2')
plt.title('Region a zakup roweru', fontsize=14, fontweight='bold')
plt.xlabel('Region', fontsize=12)
plt.ylabel('Liczba osób', fontsize=12)
plt.legend(title='Zakup roweru', labels=['Nie kupił', 'Kupił'])
plt.xticks(rotation=45)
plt.grid(axis='y', alpha=0.3)
plt.show()
```

W regionie Pacific odsetek zakupów jest najwyższy (ponad 50% osób kupiło rower). W Europie zakupy stanowią około 40%, a w Ameryce Północnej około 35%.

**Interpretacja:** Region zamieszkania ma wpływ na decyzję o zakupie roweru. Może to wynikać z różnic w infrastrukturze rowerowej, klimacie lub kulturze aktywnego spędzania czasu w poszczególnych regionach.
