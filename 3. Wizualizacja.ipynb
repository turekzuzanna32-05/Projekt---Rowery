## 3. Wizualizacja

W tej sekcji przedstawiamy 7 wykresów ilustrujących zależność między cechami klientów a decyzją o zakupie roweru. Każdy wykres został opatrzony krótkim opisem i interpretacją. Zacznijmy od instalacji bibliotek i wczytania danych.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('data/sklep_rowerowy_oczyszczone.csv')
df.head()
```

**Wykres 1. Dochód a zakup roweru**

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x='Purchased Bike', y='Income', data=df, hue='Purchased Bike', palette='Set2', legend=False)
plt.title('Rozkład dochodów a zakup roweru', fontsize=14, fontweight='bold')
plt.xlabel('Zakup roweru', fontsize=12)
plt.ylabel('Dochód [USD]', fontsize=12)
plt.xticks([0, 1], ['Nie kupił', 'Kupił'])
plt.grid(axis='y', alpha=0.3)
plt.show()
```

Osoby, które kupiły rower, mają wyraźnie wyższe dochody (mediana ok. 80 000 USD) niż osoby, które nie dokonały zakupu (mediana ok. 40 000 USD). Widać również, że w grupie kupujących występuje większe zróżnicowanie dochodów.

Dochód jest istotnym czynnikiem wpływającym na decyzję o zakupie roweru. Osoby zamożniejsze częściej decydują się na ten wydatek, co sugeruje, że rower jest postrzegany jako dobro luksusowe lub inwestycja w zdrowie.

**Wykres 2. Wiek a zakup roweru**

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x='Purchased Bike', y='Age', data=df, hue='Purchased Bike', palette='Set2', legend=False)
plt.title('Rozkład wieku a zakup roweru', fontsize=14, fontweight='bold')
plt.xlabel('Zakup roweru', fontsize=12)
plt.ylabel('Wiek [lata]', fontsize=12)
plt.xticks([0, 1], ['Nie kupił', 'Kupił'])
plt.grid(axis='y', alpha=0.3)
plt.show()
```

Kupujący rowery są przeciętnie młodsi (mediana ok. 40 lat) niż osoby niekupujące (mediana ok. 48 lat). W grupie kupujących wiek jest bardziej skupiony wokół mediany, podczas gdy w grupie niekupujących występuje szersze rozproszenie.

**Interpretacja:** Młodsze osoby (30-50 lat) częściej decydują się na zakup roweru. Może to wynikać z większej świadomości zdrowotnej, aktywnego trybu życia lub mniejszej liczby zobowiązań finansowych w porównaniu do starszych osób.

**Wykres 3. Płeć a zakup roweru**

```python
plt.figure(figsize=(8, 6))
sns.countplot(x='Gender', hue='Purchased Bike', data=df, palette='Set2', hue_order=['No', 'Yes'])
plt.title('Płeć a zakup roweru', fontsize=14, fontweight='bold')
plt.xlabel('Płeć', fontsize=12)
plt.ylabel('Liczba osób', fontsize=12)

plt.legend(
    title='Zakup roweru', 
    labels=['Nie kupił', 'Kupił'],
    loc='upper left', 
    bbox_to_anchor=(1.02, 1)
)

plt.grid(axis='y', alpha=0.3)
plt.tight_layout() 
plt.show()
```

Mężczyźni częściej kupują rowery niż kobiety. Wśród mężczyzn liczba kupujących (ok. 280) jest zbliżona do liczby niekupujących (ok. 260), podczas gdy wśród kobiet dominują osoby, które nie dokonały zakupu (ok. 290) nad kupującymi (ok. 170).

**Interpretacja:** Istnieje różnica w preferencjach zakupowych między płciami. Mężczyźni są bardziej skłonni do zakupu roweru, co może wynikać z większego zainteresowania sportem lub różnic w stylu życia.

**Wykres 4. Region a zakup roweru**

```python
plt.figure(figsize=(10, 6))
sns.countplot(x='Region', hue='Purchased Bike', data=df, palette='Set2')
plt.title('Region a zakup roweru', fontsize=14, fontweight='bold', pad=15)
plt.xlabel('Region', fontsize=12)
plt.ylabel('Liczba osób', fontsize=12)
plt.legend(title='Zakup roweru', labels=['Nie kupił', 'Kupił'], frameon=True)
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.4)
plt.show()
```

W regionie Pacific odsetek zakupów jest najwyższy (ponad 50% osób kupiło rower). W Europie zakupy stanowią około 40%, a w Ameryce Północnej około 35%.

Region zamieszkania ma wpływ na decyzję o zakupie roweru. Może to wynikać z różnic w infrastrukturze rowerowej, klimacie lub kulturze aktywnego spędzania czasu w poszczególnych regionach.

**Wykres 5. Status Związkowy a zakup roweru**

```python
plt.figure(figsize=(8, 5))

ax = sns.countplot(
    x='Marital Status', 
    hue='Purchased Bike', 
    data=df, 
    palette=['#B0BEC5', '#1E88E5'], 
    hue_order=['No', 'Yes']
)

plt.title('Stan cywilny a zakup roweru', fontsize=14, fontweight='bold', pad=15)
plt.xlabel('Stan cywilny', fontsize=12)
plt.ylabel('Liczba osób', fontsize=12)
plt.legend(title='Zakup roweru', labels=['Nie kupił', 'Kupił'])
plt.grid(axis='y', linestyle='--', alpha=0.3)
sns.despine()

plt.tight_layout()
plt.show()
```

Widzimy na wykresie, że w przypadku osób w małżeństwie więcej osób nie zakupiło roweru, niż kupiło. W przypadku singli proporcje się odwracają. 

**Wykres 6. Wielowymiarowa analiza**

Tutaj użyjemy funkcji pairplot, aby zrobic macierz wykresów.

```python
kolumny_do_wykresu = ['Age', 'Income', 'Children', 'Cars', 'Purchased Bike']

sns.pairplot(
    df[kolumny_do_wykresu], 
    hue='Purchased Bike', 
    palette=['#B0BEC5', '#1E88E5'], 
    hue_order=['No', 'Yes'],
    corner=True, 
    plot_kws={'alpha': 0.7} 
)

plt.suptitle('Wielowymiarowa analiza numeryczna: Kto kupuje rowery?', y=1.02, fontsize=16, fontweight='bold')
plt.show()
```


**1. Profil demograficzny (Wiek i Dochód)**
Analiza rozkładów brzegowych wskazuje, że decyzję o zakupie roweru najczęściej podejmują osoby w wieku średnim (ok. 35-50 lat) dysponujące dochodem na poziomie średnim i wyższym. Na wykresie punktowym relacji wieku do dochodu nie obserwuje się wyraźnego trendu liniowego, co świadczy o braku silnej współliniowości. 

**2. Charakterystyka gospodarstwa domowego (Samochody i Dzieci)**
Zagęszczenie punktów na wykresach dwuwymiarowych oraz analiza rozkładów gęstości pokazują odwrotną zależność między liczbą posiadanych samochodów i dzieci a skłonnością do zakupu roweru. Największa koncentracja kupujących występuje w grupie osób posiadających maksymalnie jeden samochód. W ujęciu wielowymiarowym zmienne reprezentujące wiek, dochód oraz liczbę samochodów wykazują najwyższą zdolność do separacji obu grup.

**Wykres 7. Heatmapa**

```python

df_corr = df.copy()

df_corr['Purchased Bike'] = df_corr['Purchased Bike'].map({'Yes': 1, 'No': 0})

kolumny_numeryczne = ['Age', 'Income', 'Children', 'Cars', 'Purchased Bike']
korelacje = df_corr[kolumny_numeryczne].corr()

plt.figure(figsize=(8, 6))
sns.heatmap(
    korelacje, 
    annot=True,          
    cmap='coolwarm',     
    fmt=".2f",           
    linewidths=0.5,      
    vmin=-1, vmax=1      
)

plt.title('Macierz korelacji: Cechy numeryczne a decyzja o zakupie', fontsize=14, fontweight='bold', pad=15)
plt.show()
```

Ogólna siła korelacji liniowych w badanym zbiorze danych jest niska. Nie obserwujemy żadnych silnych zależności między analizowanymi zmiennymi, a już napewno między zakupem roweru a pozostałymi zmiennymi.

Oznacza to brak wyraźnego, bezpośredniego i proporcjonalnego (liniowego) wpływu pojedynczych cech numerycznych na decyzję o zakupie roweru. Wszystkie widoczne na mapie słabsze powiązania należy w tym ujęciu traktować jako mało znaczące.

Widzimy umiarkowaną dodatnią korelację między Dochodem, a liczbą posiadanych aut (co ma sens logiczny) na poziomie 0.43, a także korelację 0.53 między liczbą posiadanych dzieci a wiekiem (co też trudno odnieść do samego zakupu roweru.)
